# <font color="orange">Primeiro exemplo</font>

```python
# Importamos as nossas bibliotecas necessárias

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
```

```python
# Carregamos nossos Dados

x, y = make_classification(n_samples = 200)
```

```python
x_df = pd.DataFrame(x)
x_df.head()
```

```python
y_df = pd.DataFrame(y)
y_df.head()
```

```python
x.shape, y.shape
```

```python
type(x)
```

```python
# Lembrar:
xx = [[1, 2], [3,4],[6 ,7]]

xx = np.array(xx)
xx.shape

type(xx)

xx[:,0]
xx[:,1]
```

```python
#plt.scatter(x[:,16], x[:,5], c = y)
plt.scatter(x[:,0], x[:,1], c = y)
plt.grid()
plt.show()
```

```python
# Separamos nossos Dados em Dados de treino e Dados de teste

x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
# Graficamos nossos Dados de treinamento

plt.scatter(x_train[:,0], x_train[:,1], c = y_train)
plt.title("Dados de treinamento")
plt.grid()
plt.show()

# roxo --> classe 0
# amarelo --> classe 1
```

```python
# Graficamos nossos Dados de teste

plt.scatter(x_test[:,0], x_test[:,1])
plt.title("Dados de teste")
plt.grid()
plt.show()

# Obs: esses pontos não sabemos a que classe pertencem
```

```python
# Vamos graficar os Dados juntos (tanto os Dados de treino como os Dados de teste)

plt.scatter(x_train[:,0], x_train[:,1], c = y_train)
plt.scatter(x_test[:,0], x_test[:,1])
plt.title("Dados de treinamento e Dados de teste (Juntos)")
plt.grid()
plt.show()
```

```python
# Instanciamos nosso classificador

clf = KNeighborsClassifier() # Por default número de vizinhos é 5
```

```python
# Treinamos nosso classificador

clf.fit(x_train, y_train)
```

```python
# Accuracy

clf.score(x_test, y_test)
```

```python
# Predições (vamos a predezir as labels dos pontos azuis-claros)

y_pred = clf.predict(x_test) 
y_pred
```

```python
# Etiquetas conhecidas

y_test
```

```python
matriz_confusion = confusion_matrix(y_test, y_pred)
matriz_confusion
```

```python
# Vamos apenas a observar o primeiro valor de y_pred

plt.scatter(x_train[:,0], x_train[:,1], c = y_train)
plt.scatter(x_test[0,0], x_test[0,1], s=100 ) # Nesta linha verificamos apenas um Dado (aquele ponto de cor azul em  nosso gráfico)
plt.title("Dados")
plt.grid()
plt.show()
```

```python
# Testamos com o segundo valor

plt.scatter(x_train[:,0], x_train[:,1], c = y_train)
plt.scatter(x_test[1,0], x_test[1,1], s=100 ) # Nesta linha verificamos com o segundo dado
plt.title("Dados")
plt.grid()
plt.show()
```

```python
# Carregamos as nossas Bibliotecas

import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
```

```python
# Carregamos nossos Dados de Wine

wine = load_wine()
```

```python

wine.data.shape, wine.target.shape
```

```python
# Os nomes de minhas features são:

nomes = wine.feature_names
nomes
```

```python
# Nossas features são:

wine.data

wine_data_df = pd.DataFrame(wine.data)
wine_data_df.set_axis(nomes, axis='columns', inplace=True)
wine_data_df.head()
```

```python
wine_data_df.shape
```

```python
# Visualizamos as nossas etiquetas

wine.target
```

```python
# Seus nomes são:

wine.target_names
```

```python
n_classes = len(np.unique(wine.target))

print("Número de classes:", n_classes)
```

```python
# Dados de treinamento e Dados de teste

x_train, x_test, y_train, y_test = train_test_split(wine.data, wine.target, shuffle=True)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Para este segundo exemplo também usamos o seguinte classificador


clf = KNeighborsClassifier()
```

```python

clf.fit(x_train, y_train)
```

```python
# Accuracy

clf.score(x_test, y_test)
```

```python
y_pred = clf.predict(x_test)

print(y_pred)
```

```python
# Matrix confusion 

matriz_confusion = confusion_matrix(y_test, y_pred)

print(matriz_confusion)
```
